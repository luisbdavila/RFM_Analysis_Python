We only have dates of the year 2023, so we can transform the month into numbers of days to calculate the distance to the day 1 of May (which is the day 121 of the year).


```python
# Corresponding the days completed until the start of each month, to simplify the calculation of the recency
months = {1 : 0, 2 : 31, 3 : 59, 4 : 90, 5 : 120}

# Creation of the dictionary with the information of every student
new_d = {}
# Creation of the dictionary of the items.
d_items = {}

# Creation of the dictionary of the quantity and sales per month
quantity_per_month = {"January": 0, "February": 0, "March": 0, "April": 0}
sales_per_month = {"January": 0, "February": 0, "March": 0, "April": 0}

# We go through every dictionary in the dataset
for d in dataset:

    # If we don’t have the name, we introduce the key 
    if d['Name'] not in new_d.keys():
        new_d[d['Name']] = {}
        new_d[d['Name']]['Dates'] = [d['Date'][0] + months[d['Date'][1]]] # Number of the day + Number of days in the completed months
        new_d[d['Name']]['Money Spent'] = d['Unit Price'] * d['Quantity'] 
    # If we already have the name, we update the values.
    else:
        new_d[d['Name']]['Dates'].append(d['Date'][0] + months[d['Date'][1]])
        new_d[d['Name']]['Money Spent'] += d['Unit Price'] * d['Quantity']

    # If we don’t have the item, we introduce the key.
    if d['Item'] not in d_items.keys():
        d_items[d['Item']] = {}
        d_items[d['Item']]['Number of items sold'] = d['Quantity']
        d_items[d['Item']]['Total sales'] = d['Unit Price'] * d['Quantity']
    # If we have the item, we update the values.    
    else:
        d_items[d['Item']]['Number of items sold'] += d['Quantity']
        d_items[d['Item']]['Total sales'] += d['Unit Price'] * d['Quantity']

    # If is january 
    if d["Date"][1]==1:
        quantity_per_month["January"]+= d['Quantity']
        sales_per_month["January"]+= d['Unit Price'] * d['Quantity']
    # If is february 
    elif d["Date"][1]==2:
        quantity_per_month["February"]+= d['Quantity']
        sales_per_month["February"]+= d['Unit Price'] * d['Quantity']
    # If is march 
    elif d["Date"][1]==3:
        quantity_per_month["March"]+= d['Quantity']
        sales_per_month["March"]+= d['Unit Price'] * d['Quantity']
    # If is april
    elif d["Date"][1]==4:
        quantity_per_month["April"]+=  d['Quantity']
        sales_per_month["April"]+= d['Unit Price'] * d['Quantity']

# List to have the corresponding recency, frequency and monetary (tuple of the value and the index).
Frequency = []
Last_Purchases = []
Monetary = []
# List to be used in the histogram.
A_elements = []

# We go through every name in the new dictionary
for i, key in enumerate(new_d.keys()):

    # The frequency will be the number of dates (number of times he/she bought something).
    new_d[key]['Frequency'] = len(new_d[key]['Dates'])
    # The last purchase/recency will be the first day of May (day 121) - the most recent purchase
    # Thus, the recency is the number of days since the last purchase.
    new_d[key]['Last Purchase'] = 121 - max(new_d[key]['Dates'])
    # The money spent is the same value we had, we only round it
    new_d[key]['Money Spent'] = round(new_d[key]['Money Spent'],2)
    # The average money is the money spent divided by the frequency.
    new_d[key]['Average'] = round(new_d[key]['Money Spent'] / new_d[key]['Frequency'],2)
    # We delete the date because it will no longer be useful.
    del new_d[key]['Dates']
    # We create a place to store the RFM and RFM' in a form of list (RFMP stands for RFM Prime or RFM').
    new_d[key]['RFM'] = []
    new_d[key]['RFMP'] = []

    # These are the lists we will use to sort in the RFM calulation.
    Frequency.append( (new_d[key]['Frequency'] , i) )
    Last_Purchases.append( (new_d[key]['Last Purchase'], i) )
    Monetary.append( (new_d[key]['Money Spent'], i ) )
    # These are the list we will use to plot the histograms.
    A_elements.append(new_d[key]['Average'])


print(f'This is an example of how our dictionary looks, every key is a name and the value is the characteristics of that student: \n Carolina :{new_d["Carolina"]}')
#This is an example of how our dictionary looks, every key is a name and the value is the characteristics of that student:
```

# Chapter 3: Sorting

### First RFM scores (RFM)

The RFM consists in sorting first the recency and giving a value of 3 to 1 (3 are the customers that have more recent purchases), then sorting the frequency and giving a value of 3 to 1 (3 are the customers that have the highest frequency), and finally do the same for the monetary (3 are the customers that spent the most money).

To do so we create a function that uses merge sort and can sort in ascending or descending order whether you introduce 0 or 1.

We will us this merge sort function in all our algorithm because we tried to use and algorithm that was the faster regarding the average and worst possible scenario (because if we use quick sort, we could encounter a case when the pivot was the last or first element and the time complexity will increase dramatically; this is in the case of the recency where the majority of the values are the minimum), also in the sorting of the frequency and the monetary we don’t want to change the order of that list, for that reason we need to create a separate list that is sorted (and still have the list unsorted to compute more efficiently the RFM'). We believe this overhead won’t be a problem because the number of customers is small.


```python
# A is in the form of a list of tuples with the first element to be sorted (the actual value) and the second to be the index of the original list.
# The index will help us locate the name in the dictionary (the index is unique to every person).
# This sorting works because: (20,0) > (15,10) = True
# This also works with names : 'Dinis' < 'Luis' = True (both with capital letter in the first letter as our dataset).
# This also works with normal numbers.

def sort_merge(A, option=0): # 0 for bigest to smallest , 1 for smallest to biggest (If you don’t put anything it will sort biggest to smallest).
    # If the list is only 1 value we will return the value
    if len(A) == 1:
        return A
    
    # We divide the list in 2 
    middle = len(A)//2
    A_left = sort_merge(A[:len(A)//2],option=option)
    A_right = sort_merge(A[len(A)//2:],option=option)
    
    #We merge the lists in an ordered
    return merge_d(A_right , A_left, option)

def merge_d(a_right , a_left, option):
    # The sorted list of the 2 arrays
    a_sorted = []
    i_right , i_left = 0 , 0

    while i_right < len(a_right) and i_left < len(a_left):

        # If the option is 0 we will sort biggest to smallest (descending order)
        if option==0:
            # We append the biggest value
            if a_right[i_right] > a_left[i_left]:
                a_sorted.append(a_right[i_right])
                i_right += 1
            else:
                a_sorted.append(a_left[i_left])
                i_left += 1
                
        # If the option is 1 we will sort smallest to biggest (ascending order)
        else:
            # We append the smallest value
            if a_right[i_right] < a_left[i_left]:
                a_sorted.append(a_right[i_right])
                i_right += 1
            else:
                a_sorted.append(a_left[i_left])
                i_left += 1
                
    # We will complete the final array with the elements that are left
    # If we finish the right part, we will append the left part (and the opposite in the else)
    if i_right == len(a_right):
        a_sorted.extend(a_left[i_left:])
    
    else:
        a_sorted.extend(a_right[i_right:]) 
    
    # This is the array sorted
    return a_sorted

```

We created a function to append the respective value of the recency, frequency and monetary, given the sorted array (3 in the first third, 2 in the second, 1 in the last); and it also works in both scores (the option = 1 computes RFM, and the option = 0 computes RFM'(RFMP)).

```python
def RFM_number(L_sorted, Dic_names, option=1):
    # If the option is 1 we will append it to RFM (we can leave the option blank and we still get the RFM score)
    if option == 1:
        key = 'RFM'
    # If the option is 0 we will append to RFMP
    else:
        key = 'RFMP'
    
    # Name of every person
    names = list(Dic_names.keys())

    # Dividing the sorted (descending order) list in 3 
    division = len(L_sorted)/3 
    start = 0
    # Go through every element in the sorted array
    for elements in L_sorted:
      
        # Give 3 to the first third of the array
        if start < int(division):
            Dic_names[names[elements[1]]][key].append(3)
        # Give 2 to the second third of the array    
        elif start < int(2*division):
            Dic_names[names[elements[1]]][key].append(2)
        # Give 1 to the last third of the array
        else:
            Dic_names[names[elements[1]]][key].append(1)
        
        start +=1
    
```

Because our function RFM_number appends the value it is important to have the correct order when we run our algorithm (we need to assign  the number of the recency first, then the frequency and finally the monetary).

```python
# We sort our arrays of recency, frequency and monetary
sorted_r = sort_merge(Last_Purchases,1)
sorted_f = sort_merge(Frequency)
sorted_m = sort_merge(Monetary)

# We compute the score of recency, frequency and monetary in the RFM (in order).
RFM_number(sorted_r, new_d)
RFM_number(sorted_f, new_d)
RFM_number(sorted_m, new_d)
```

### Second RFM Scores (RFM')

The second score (RFM') consists in sorting by the recency, giving values (3,2,1) like the normal RFM score, then dividing in 3 ‘buckets´ and sorting their frequencies (within each bucket) and giving values (3,2,1) like the normal RFM score but within each "bucket", and then in each bucket (9) of the frequency we sort by the monetary and give values (3,2,1).

To do so we create a function that computes the score, it receives as parameters (guide = sorted list of recency, match = list of frequency, end = list of monetary) it has a parameter called step, but the user doesn’t need to set it because it is supposed to start at 0.

We don’t sort recency again inside the function because we already have done it in the first score.

```python
def RFMP (guide, mach, end, Dic_names, step=0):
    # When it's the second and third interaction the step is 1 and 2 (they are different from 0)
    if step != 0:
        # Create a list that will have the values of the frequency or monetary of the respective bucket
        l_match = []
        for elements in guide:
            # Append only the elements of the bucket (the elements corresponding to the persons inside guide).
            l_match.append(mach[elements[1]])
        
        # Sort the list to compute the respective RFM' value
        M_sorted= sort_merge(l_match)
        # Give the values of the sorted match in the RFMP 
        RFM_number(M_sorted,Dic_names,0)
        
        # if you already have the values of the frequencies in the RFM' now your next match will be the monetary (end)
        if step == 1:
            mach = end
            
    # When we start, step = 0, so it give the values of the sorted recency in the RFMP        
    else:
        M_sorted = guide
        RFM_number(guide,Dic_names,0)
    
    # Steps count how many parts the RFM' have been computed
    step += 1
    # There are 3 lists, steps are 0,1,2, when you reach 3 you have finished
    if step == 3:
        return None
    
    # Divide the guide into 3 
    division = len(M_sorted)/3
    
    # You want to order the next list in relation to the previous one, and repeat the process 
    # First part
    A_3 = M_sorted[:int(division)]
    RFMP( A_3, mach, end, Dic_names, step)
    # Second part
    A_2 = M_sorted[int(division) :int(division*2)]
    RFMP(A_2, mach, end, Dic_names, step)
    # Final part
    A_1 = M_sorted[int(division*2):]
    RFMP(A_1, mach, end, Dic_names, step)
```

### Binary Search

```python
# This searches for a name in the dictionary using binary search (better time complexity than linear search)
def binary_search_string(l, name, start, end):
    # If the start index is higher the end index it means that the name doesn't exist in our dataset
    if start > end:
        return f'We have no information about {name}'
    
    else:
        # We will look for the name in the middle
        num = (start + end) // 2   
        # If  it is the name were looking for we will return the score
        if l[num] == name:
            return new_d[name]['RFM']
        # If the name is 'lower' (comes before the middle name) we will look in the first half
        elif l[num] > name:
            # We repeat the process recursively until we find a solution (the scores or that the person we're looking for is not in the dataset)
            return binary_search_string(l, name, start , num-1)
        # If the name is 'higher' (comes after the middle name) we will look in the second half
        else:
            # We repeat the process recursively until we find a solution (the scores or that the person we're looking for is not in the dataset)
            return binary_search_string(l, name , num+1 , end) 
            
# This function allows the user to only enter the dictionary and the name they want to look for using the binary search     
def binary_search_RFM(d, name):
    return binary_search_string(list(d.keys()), name , 0 , len(list(d.keys()))-1 )
```
